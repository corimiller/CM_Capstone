```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

X = train_set[['PercentWomen']]
y = train_set['GradRate']

X_test = test_set[['PercentWomen']]
y_test = test_set['GradRate']

tree_model = DecisionTreeRegressor()
tree_model.fit(X, y)

y_pred = tree_model.predict(X)
print('Results for decision tree on training data')
print('Default settings')
print('Mean Squared Error:', mean_squared_error(y, y_pred))
print('R-squared:', r2_score(y, y_pred))

y_test_pred = tree_model.predict(X_test)
print('Results for decision tree on test data')
print('Default settings')
print('Mean Squared Error:', mean_squared_error(y_test, y_test_pred))
print('R-squared:', r2_score(y_test, y_test_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

X = train_set[['PercentWomen']]
y = train_set['GradRate']

X_test = test_set[['PercentWomen']]
y_test = test_set['GradRate']

rf_model = RandomForestRegressor()
rf_model.fit(X, y)

y_pred = rf_model.predict(X)
print('Results for random forest regressor on training data')
print('Default settings')
print('Mean Squared Error:', mean_squared_error(y, y_pred))
print('R-squared:', r2_score(y, y_pred))

y_test_pred = rf_model.predict(X_test)
print('Results for random forest regressor on test data')
print('Default settings')
print('Mean Squared Error:', mean_squared_error(y_test, y_test_pred))
print('R-squared:', r2_score(y_test, y_test_pred))
```
